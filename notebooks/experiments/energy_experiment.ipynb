
# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import sys, os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../../")))
from src.predictions.estimator import train_energy_model, generate_training_data
from src.anomaly.detector import train_anomaly_detector, flag_anomaly
```

# Load & Generate Data

```python
X, y = generate_training_data()
model = train_energy_model()
```

# Predictions

```python
pred = model.predict(X)
plt.scatter(y, pred, alpha=0.7)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel("Actual Energy (kWh)")
plt.ylabel("Predicted Energy (kWh)")
plt.title("Predicted vs Actual Energy")
plt.show()
```
